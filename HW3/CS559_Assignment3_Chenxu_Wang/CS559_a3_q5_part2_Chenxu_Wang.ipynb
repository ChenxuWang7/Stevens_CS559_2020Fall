```python
from sklearn.datasets import load_iris
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelBinarizer
import matplotlib.pyplot as plt
```

```python
# TODO get x(data) and y(label) from iris
iris = load_iris()
X = iris.data  # 150x4
y = iris.target # 150x1
```

```python
# One hot coidng for 3 classification
one_hot_y =LabelBinarizer().fit_transform(y)
```

```python
# # TODO: Split data into train(80%) and test(20%) set. 
train_X, test_X, train_y, test_y = train_test_split(X,one_hot_y, test_size=0.2,random_state=42)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
# Build the model
model = Sequential()
model.add(Dense(8, input_shape=(4,), activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(3, activation='softmax', name='output'))

# Use Adam optimizer with learning rate of 0.001
optimizer = Adam(lr=0.001)
model.compile(optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
# Train the model
max_iterations = 500
history = model.fit(train_X, train_y, verbose=2, batch_size=10, epochs=max_iterations)
```

```python
acc=history.history['accuracy']

plt.plot(range(1,max_iterations+1),acc, '-.r', LineWidth=2,label='train_accuracy')
plt.xlabel('iterations')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
results = model.evaluate(test_X, test_y)

print('Test Accuracy: {}%'.format(results[1]*100))
print('Test loss: {:4f}'.format(results[0]))
```
